```python
a=[]

for i in range(len(df['ActivityDate'].unique())):
    x=0
    for j in range(len(df['ActivityDate'])):
        if df['ActivityDate'].unique()[i]==df['ActivityDate'][j]:
            x+=df['TotalSteps'][j]
    a.append(x)

m=[v for v in df['ActivityDate'].unique()]              
```

```python
b=pd.DataFrame({'date':m,'totalstep':a})
```

```python
c=b.head(20)
c
```
